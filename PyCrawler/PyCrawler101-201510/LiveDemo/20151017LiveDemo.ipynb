# map

```python
map(lambda e:e.attrib["href"],doc.xpath("//ul[@class='list_1 bd_dbottom']//li/a"))
```

# iterator

```python
[e.attrib["href"] for e in doc.xpath("//ul[@class='list_1 bd_dbottom']//li/a")]
```
